```python
import matplotlib.pyplot as plt
import numpy as np
from wordcloud import WordCloud

# Create a DataFrame with country counts
title = pd.DataFrame()
title['name'] = athletes_df['country'].value_counts()

# Convert the DataFrame to a dictionary
title_dict = title['name'].to_dict()

# Create a WordCloud object with custom colors
color_map = plt.cm.get_cmap('tab20')  # Choose a colorful colormap
wc = WordCloud(
    width=1000,
    height=600,
    max_words=20,
    colormap=color_map,
    contour_width=0,  # Set the contour width to 0 to remove the border
    prefer_horizontal=0.8  # Adjust the orientation of the words
)

# Generate the WordCloud from the frequencies
wc.generate_from_frequencies(title_dict)

# Set a custom color function for the background
def gradient_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    return tuple(np.random.randint(0, 256, 3))  # Generate random RGB values

# Set the background color to a gradient black effect
wc.recolor(color_func=gradient_color_func, random_state=42)

# Plot the WordCloud with a black background
plt.figure(figsize=(10, 10))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.tight_layout()
plt.gca().set_facecolor('black')  # Set the background color of the plot to black
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from wordcloud import WordCloud
import base64

# Create a DataFrame with country counts
title = pd.DataFrame()
title['name'] = athletes_df['country'].value_counts()

# Convert the DataFrame to a dictionary
title_dict = title['name'].to_dict()

# Create a WordCloud object with custom colors
color_map = plt.cm.get_cmap('tab20')  # Choose a colorful colormap
wc = WordCloud(
    width=1000,
    height=600,
    max_words=20,
    colormap=color_map,
    contour_width=0,  # Set the contour width to 0 to remove the border
    prefer_horizontal=0.8  # Adjust the orientation of the words
)

# Generate the WordCloud from the frequencies
wc.generate_from_frequencies(title_dict)

# Set a custom color function for the background
def gradient_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    return tuple(np.random.randint(0, 256, 3))  # Generate random RGB values

# Set the background color to a gradient black effect
wc.recolor(color_func=gradient_color_func, random_state=42)

# Save the WordCloud as an image
wc_image_path = 'wordcloud.png'
wc.to_file(wc_image_path)

# Encode the WordCloud image to base64
with open(wc_image_path, 'rb') as f:
    image_data = base64.b64encode(f.read()).decode('utf-8')

# Create the HTML content to display the WordCloud image
html_content = f'''
<html>
<head>
  <title>WordCloud</title>
</head>
<body>
  <div align="center">
    <h1>WordCloud</h1>
    <img src="data:image/png;base64,{image_data}" alt="WordCloud">
  </div>
</body>
</html>
'''

# Save the HTML content to a file
html_file_path = 'wordcloud.html'
with open(html_file_path, 'w') as f:
    f.write(html_content)

# Open the HTML file in the default web browser
import webbrowser
webbrowser.open(html_file_path)
```

```python
import matplotlib.pyplot as plt

# Create a DataFrame with country counts
title = pd.DataFrame()
title['name'] = athletes_df['country'].value_counts()

# Calculate the total count of athletes
total_count = title['name'].sum()

# Calculate the percentage for each country
title['percentage'] = (title['name'] / total_count) * 100

# Sort the DataFrame by count in descending order
title = title.sort_values('name', ascending=False)

# Plot the bar graph
plt.figure(figsize=(10, 6))
plt.bar(title.index, title['percentage'], color='skyblue')

# Add percentage labels to the bars
for i, percentage in enumerate(title['percentage']):
    plt.text(i, percentage + 1, f"{percentage:.1f}%", ha='center', va='bottom')

# Customize the plot
plt.xlabel('Country')
plt.ylabel('Percentage')
plt.title('Athletes from Different Countries')
plt.xticks(rotation=90)
plt.ylim(0, 100)
plt.grid(axis='y')

plt.tight_layout()
plt.show()
```
