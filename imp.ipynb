```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
Title = pd.DataFrame()

Title['name'] = athletes_df['country'].value_counts() 
Title = dict(zip (Title.index, Title['name'].tolist()))

wc = WordCloud(background_color='white', width=1000, height=600, max_words=20).generate_from_frequencies(Title)

plt.figure(figsize=(10, 10))

plt.imshow(wc, interpolation='bilinear')

plt.axis('off') 
plt.show()
```
